## # Extraindo informações com dados socioeconômicos

Nota da Redação (NU_NOTA_REDACAO), e os Questionários sobre Pai e Mãe (Q001 e Q002)


- Excluindo linhas sem a nota de Matemática e Redação (dropna)


- Criando dicionários com as respostas dos Questionários

Video: https://www.youtube.com/watch?v=ZcvFSEaRcTg

```python
# SELECIONANDO AS COLUNAS PARA O NOVO DATAFRAME

colunas_dados_socioeconomicos = ['NU_INSCRICAO', 'NU_NOTA_MT', 'NU_NOTA_REDACAO', 'Q001', 'Q002']
```

```python
# CRIANDO NOVO DATAFRAME COM AS COLUNAS SELECIONADAS DO DATAFRAME ORIGINAL (enem2018)

dados_socioeconomicos = enem2018.filter(items=colunas_dados_socioeconomicos)
dados_socioeconomicos

# Total de linhas: 100001
```

## # Extraindo informações com dados socioeconômicos usando WHERE
Usando as colunas Nº de Inscrição (NU_INSCRICAO), Nota de Matemática (NU_NOTA_MT).
Nota da Redação (NU_NOTA_REDACAO), os Questionários sobre Pai e Mãe (Q001 e Q002)
e o Estado de residência (SG_UF_RESIDENCIA)

Video: https://www.youtube.com/watch?v=RXbX6giU2vI

```python
# EXIBINDO O CABEÇALHO DO DATAFRAME 'dados_socioeconomicos'
dados_socioeconomicos
```

```python
# ADICIONANDO A COLUNA 'SG_UF_RESIDENCIA' AO DATAFRAME 'dados_socioeconomicos'

dados_socioeconomicos['SG_UF_RESIDENCIA'] = enem2018['SG_UF_RESIDENCIA']
```

```python
# ADICIONANDO A COLUNA 'SG_UF_RESIDENCIA' AO DATAFRAME 'dados_socioeconomicos' USANDO .loc

dados_socioeconomicos.loc[0:99999]['SG_UF_RESIDENCIA'] = enem2018['SG_UF_RESIDENCIA']
dados_socioeconomicos
```

```python
# FAZENDO A MÉDIA ENTRE A NOTA DA REDAÇÃO E A RESPOSTA DO QUESTIONÁRIO NO_Q002

dados_socioeconomicos.filter(items=['NU_NOTA_REDACAO','NO_Q002'])\
                     .groupby('NO_Q002')\
                     .mean().sort_values(by='NU_NOTA_REDACAO', ascending=False)
```

```python
# FAZENDO A MÉDIA ENTRE A NOTA DA REDAÇÃO E A RESPOSTA DO QUESTIONÁRIO NO_Q002 DA PARAÍBA

dados_socioeconomicos.filter(items=['NU_NOTA_REDACAO','NO_Q002'])\
                     .where(dados_socioeconomicos.SG_UF_RESIDENCIA == 'RN')\
                     .groupby('NO_Q002')\
                     .mean().sort_values(by='NU_NOTA_REDACAO', ascending=False)
```

```python
# EXTRAINDO A MÉDIA ENTRE A NOTA DA REDAÇÃO ONDE A MÃE DO PARTICIPANTE 'Completou a Pós-graduação.'
# MÉDIA: DA MENOR PARA A MAIOR

dados_socioeconomicos.filter(items=['SG_UF_RESIDENCIA','NU_NOTA_REDACAO','NO_Q002'])\
                     .where(dados_socioeconomicos.NO_Q002 == 'Completou a Pós-graduação.')\
                     .groupby(['SG_UF_RESIDENCIA','NO_Q002'])\
                     .mean().sort_values(by='NU_NOTA_REDACAO', ascending=True)
```

```python
# EXTRAINDO A MÉDIA ENTRE A NOTA DA REDAÇÃO ONDE A MÃE DO PARTICIPANTE 'Nunca estudou.'
# MÉDIA: DA MAIOR PARA A MENOR
# RESULTADO: A MENOR MÉDIA GERAL (AC: 554.54) ONDE A MÃE DO PARTICIPANTE 'Completou a Pós-graduação.'
# AINDA É SUPERIOR A MAIOR GERAL (AC: 496.15) ONDE A MÃE DO PARTICIPANTE 'Nunca estudou.'

dados_socioeconomicos.filter(items=['SG_UF_RESIDENCIA','NU_NOTA_REDACAO','NO_Q002'])\
                     .where(dados_socioeconomicos.NO_Q002 == 'Nunca estudou.')\
                     .groupby(['SG_UF_RESIDENCIA','NO_Q002'])\
                     .mean().sort_values(by='NU_NOTA_REDACAO', ascending=False)
```

```python
# GERANDO O GRÁFICO DA MÉDIA ENTRE A NOTA DA REDAÇÃO E A RESPOSTA DO QUESTIONÁRIO NO_Q002 DOS ESTADOS

dados_socioeconomicos.filter(items=['SG_UF_RESIDENCIA','Q002','NU_NOTA_REDACAO'])\
                     .groupby(['Q002','SG_UF_RESIDENCIA'])\
                     .mean().sort_values(by='NU_NOTA_REDACAO', ascending=False)\
                     .unstack().plot()
```

```python
# EDITANDO O GRÁFICO
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(25,13))
plt.suptitle('Nota Redação x Escolaridade x Estado', size=30)

dados_socioeconomicos.filter(items=['SG_UF_RESIDENCIA','Q002','NU_NOTA_REDACAO'])\
                     .groupby(['Q002','SG_UF_RESIDENCIA'])\
                     .mean().sort_values(by='NU_NOTA_REDACAO', ascending=False)\
                     .unstack().plot(ax=ax)
```
